```python
#Exploring the mock catalog
```

```python
from astropy.table import Table
import numpy as np
t_mock=Table.read('data/second_catalog.fits')
print(t_mock)
```

```python
print(t_mock.colnames)
```

```python
#removing zero values from the mock for redshift and theta core
t_mock['REDSHIFT'][t_mock['REDSHIFT'] == 0] = 0.01
t_mock['THETA_CORE'][t_mock['THETA_CORE'] == 0] = 0.01
```

```python
#Exploring the real one
```

```python
t_real=Table.read('data/SPT_SZ_pol_230420.fits')
print(t_real)
print(t_real["REDSHIFT"])
```

```python
print(t_real.colnames)
```

```python
#Let's create a new table that has the same format as the real one but when the same column name occours we put the mock data instead of the real one
```

```python
import random
t_new=t_real
col_t_mock=t_mock.colnames
col_t_new=t_real.colnames
idx = random.sample(range(len(t_mock)),len(t_real))
banned=["SPT_ID"]
#NOT SHUFFLED
r=np.arange(0,len(t_real))
for col in col_t_new:
    if col in col_t_mock:
        if col not in banned:
            t_new[col]=t_mock[col][r]
    else:
        continue
        
#t_new is the new catalog
```

```python
t_new.write('data/cat_from_mock.fits', overwrite=True)
```

```python
t_file=Table.read('data/cat_from_mock.fits')
print(t_file["REDSHIFT"])
```
